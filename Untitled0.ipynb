```python
import torch
import torch.nn as nn

class UniversalDummyModel(nn.Module):
    def __init__(self):
        super().__init__()

        self.net = nn.Sequential(
            nn.LazyLinear(128),  # otomatik
            nn.ReLU(),
            nn.Linear(128, 128), # Arakatman
            nn.ReLU(),
            nn.Linear(128, 13),  # ÇIKIŞ: Robotun 13 motoruna giden sinyal
            nn.Tanh()            # Sinyali -1 ile +1 arasına sıkıştır
        )

    def forward(self, x):

        if isinstance(x, dict):

            x = torch.cat([v.flatten() for v in x.values()], dim=-1)


        if len(x.shape) == 1:
            x = x.unsqueeze(0)

        return self.net(x)


model = UniversalDummyModel()


dummy_input = torch.randn(1, 50)


model(dummy_input)


try:
    traced_model = torch.jit.trace(model, dummy_input)

    traced_model.save("konnex_universal_model.pt")
    print("BAŞARILI: 'konnex_universal_model.pt' dosyası oluşturuldu.")
    print("Bu dosya her ortamda çalışır. Sol menüden indirip GitHub'a yükleyebilirsin.")
except Exception as e:
    print(f"Hata oluştu: {e}")
```
